```python
#2
colors =['r','y','b','g','black']
plt.subplot(331)
plt.barh(cal2.index, cal2['total_bedrooms'],color = colors)
plt.title('total_bedrooms')
plt.ylabel('ocean_proximity')
plt.subplot(333)
plt.barh(cal2.index, cal2['population'],color = colors)
plt.title('population')
plt.ylabel('ocean_proximity')
plt.subplot(337)
plt.barh(cal2.index, cal2['households'],color = colors)
plt.title('households')
plt.ylabel('ocean_proximity')
plt.subplot(339)
plt.barh(cal2.index, cal2['total_rooms'],color = colors)
plt.title('total_rooms')
plt.ylabel('ocean_proximity')
```

```python
##색깔을 입혀보자. 거주지역이 NEAR_BAY인 사람들과 아닌 사람들로 나눠서 보고싶다면?
cols = []
for i in cali['ocean_proximity']:
    choice = 0
    if i == 'NEAR OCEAN':
        choice = 1
        
    else:
        choice = 2
    cols.append(choice)
```

```python
plt.scatter(cali['median_income'], cali['median_house_value'], alpha=0.3, c=cols)
plt.ylabel('median_house_value')
plt.xlabel('median_income')
```
