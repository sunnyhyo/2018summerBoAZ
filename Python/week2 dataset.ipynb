> ### List 내장 함수
* list_ .append(x) -> 리스트의 맨 마지막에 x를 추가시키는 함수이다.
* list_ .sort() -> 리스트의 요소를 순서대로 정렬해 준다.
* list_ .reverse() -> 현재의 리스트를 그대로 거꾸로 뒤집어 준다.
* list_ .index(x) -> 함수는 리스트에 x라는 값이 있으면 x의 위치값을 리턴한다.
* list_ .insert(x, y) -> 리스트의 x번째 위치에 y를 삽입하는 함수이다.
* list_ .remove(x) -> 리스트에서 첫 번째로 나오는 x를 삭제한다.
* list_ .pop() -> 리스트의 맨 마지막 요소를 돌려 주고 그 요소는 삭제하는 함수이다.
* list_ .pop(x) -> 리스트의 x번째 요소를 돌려 주고 그 요소는 삭제한다.
* list_ .count(x) -> 리스트 내에 x개수를 조사해 리턴해준다.
* list_ .extend(x) -> x에는 리스트만 올 수 있으며 원래의 리스트에 x 리스트를 더하게 된다.

> ### 리스트의 문자열을 int 형태로 변환
* list_a = ['1', '2', '3', '4']   -> list_a = [1, 2, 3, 4] 로 바꾸고 싶을 때,
* list_a = list(map(int, list_a))   를 해주면 된다.
* 혹은 
* list_a = [int (i) for i in list_a]

> ### Dictionary 내장 함수
* dict_.keys() -> key만을 모아서 dict_keys라는 객체를 리턴한다.
* dict_.values() -> value만을 모아서 dict_values라는 객체를 리턴한다.
    * 위의 두 객체를 리스트로 사용하기위해선 list(객체)로 형변환해야함
* dict_.clear() -> 딕셔너리 안의 모든 요소를 삭제한다
* dict_.get(key) -> key에 대응되는 value 를 리턴해준다.
* 'key' in dict_ -> 딕셔너리에 key 가 있는지 확인
* dict_.update(x) -> x에는 딕셔너리만 올 수 있으며 원래의 딕셔너리에 x 딕셔너리를 더하게 된다.

> ### Set 특징
* 중복을 허용하지 않는다.
* 순서가 없다(Unordered).

> ### Set 내장 함수 
* set1_ & set2_ -> 교집합
* set1_.intersection(set2_) -> 교집합
* set1_ | set2_ -> 합집합
* set1_.union(set2_) -> 합집합
* set1_ - set2_ -> 차집합
* set1_.difference(set2_) -> 차집합
* set1_.add(x) -> 집합에 x 추가한다.
* set1_.update(set2_) -> s1에 s2 원소들을 추가한다.
* set1_.remove(x) -> 집합에서 x를 제거한다.
